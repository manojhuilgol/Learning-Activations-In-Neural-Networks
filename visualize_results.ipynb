```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

### **IRIS**

```python
with open('/content/drive/My Drive/code_challenges/train_and_test_metrics_norm_distribution_mean_1_std_half.pkl', 'rb') as handle:
    results_dict_1 = pickle.load(handle)


with open('/content/drive/My Drive/code_challenges/train_and_test_metrics_norm_distribution_mean_2_std_1_negative.pkl', 'rb') as handle:
    results_dict_2 = pickle.load(handle)

```

### Metrics

```python
train_acc_case_1 = np.round(results_dict_1["accuracy"]["train"][-1], 3)
train_loss_case_1 = np.round(results_dict_1["loss"]["train"][-1], 3)

test_acc_case_1 = np.round(results_dict_1["accuracy"]["test"][-1], 3)
test_loss_case_1 = np.round(results_dict_1["loss"]["test"][-1], 3)

train_acc_case_2 = np.round(results_dict_2["accuracy"]["train"][-1], 3)
train_loss_case_2 = np.round(results_dict_2["loss"]["train"][-1], 3)

test_acc_case_2 = np.round(results_dict_2["accuracy"]["test"][-1], 3)
test_loss_case_2 = np.round(results_dict_2["loss"]["test"][-1], 3)

f_score_case_1 = np.round(results_dict_1["f_measure"]["test"][-1], 3)
f_score_case_2 = np.round(results_dict_2["f_measure"]["test"][-1], 3)

print ("The Training and Test accuracy for CASE-1 = {}, {}. And the train, test loss = {}, {}. The F-measure for the test set = {}".format(train_acc_case_1, test_acc_case_1, train_loss_case_1, test_loss_case_1, f_score_case_1))
print ("The Training and Test accuracy for CASE-2 = {}, {}. And the train, test loss = {}, {}. The F-measure for the test set = {}".format(train_acc_case_2, test_acc_case_2, train_loss_case_2, test_loss_case_2, f_score_case_2))
```

```python
plt.figure(figsize=(16, 7))
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

train_acc_1 = results_dict_1["accuracy"]["train"]
train_acc_2 = results_dict_2["accuracy"]["train"]

plt.plot(train_acc_1, linewidth=2, color="blue", label="case-1")
plt.plot(train_acc_2, linewidth=2, color="green", label="case-2")

plt.title("Training Accuracy", fontweight="bold", fontsize=15)
plt.ylabel("Accuracy in %", fontweight="bold", fontsize=15)
plt.xlabel("Number of Epochs", fontweight="bold", fontsize=15)

plt.legend(fontsize=14)
plt.grid(color="brown")
plt.show()
```

```python
plt.figure(figsize=(16, 7))
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

train_loss_1 = results_dict_1["loss"]["train"]
train_loss_2 = results_dict_2["loss"]["train"]

plt.plot(train_loss_1, linewidth=2, color="blue", label="case-1")
plt.plot(train_loss_2, linewidth=2, color="green", label="case-2")

plt.title("Training Loss", fontweight="bold", fontsize=15)
plt.ylabel("Loss values", fontweight="bold", fontsize=15)
plt.xlabel("Number of Epochs", fontweight="bold", fontsize=15)

plt.legend(fontsize=14)
plt.grid(color="brown")
plt.show()
```

```python
plt.figure(figsize=(20, 7))

k0, k1 = {key: [] for key in ["K1", "K2"]}, {key: [] for key in ["K1", "K2"]}

for K1, K2 in zip(results_dict_1["K"]["train"], results_dict_2["K"]["train"]):
  k0["K1"].append(K1[0])
  k1["K1"].append(K1[1])

  k0["K2"].append(K2[0])
  k1["K2"].append(K2[1])


plt.subplot(2, 2, 1)
plt.plot(k0["K1"], linewidth=5, color="blue", label="case-1")
plt.title("k0", fontsize=13, fontweight="bold")
plt.xlabel("Number of Epochs", fontweight="bold")
plt.yticks(np.arange(min(k0["K1"]).item(), max(k0["K1"]).item(), (max(k0["K1"]).item() - min(k0["K1"]).item())/3), rotation=45)
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(k0["K2"], linewidth=5, color="blue", label="case-2")
plt.title("k0", fontsize=13, fontweight="bold")
plt.xlabel("Number of Epochs", fontweight="bold")
plt.yticks(np.arange(min(k0["K2"]).item(), max(k0["K2"]).item(), (max(k0["K2"]).item() - min(k0["K2"]).item())/3), rotation=45)
plt.grid()
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(k1["K1"], linewidth=5, color="green", label="case-1")
plt.title("k1", fontsize=13, fontweight="bold")
plt.xlabel("Number of Epochs", fontweight="bold")
plt.yticks(np.arange(min(k1["K1"]).item(), max(k1["K1"]).item(), (max(k1["K1"]).item() - min(k1["K1"]).item())/3), rotation=45)
plt.grid()
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(k1["K2"], linewidth=5, color="green", label="case-2")
plt.title("k1", fontsize=13, fontweight="bold")
plt.xlabel("Number of Epochs", fontweight="bold")
plt.yticks(np.arange(min(k1["K2"]).item(), max(k1["K2"]).item(), (max(k1["K2"]).item() - min(k1["K2"]).item())/3), rotation=45)
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
